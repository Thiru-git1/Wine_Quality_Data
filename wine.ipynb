```python
df_wine = df.copy()
```

```python
def categorize_quality(value):
    if value <= 5:
        return 'Low'
    elif value == 6:
        return 'Medium'
    else:
        return 'High'

df_wine['quality_label'] = df_wine['quality'].apply(categorize_quality)
```

```python
df_wine
```

```python
df_wine['color'] = df_wine['color'].map({'red': 0, 'white': 1})
```

```python
df_wine
```

```python
df_wine = df_wine.drop(columns=['quality'])
```

```python
df_wine
```

```python
df_wine.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_wine["quality_label"] = le.fit_transform(df_wine["quality_label"])
```

```python
df_wine.info()
```

```python
df_wine
```

```python
from sklearn.model_selection import train_test_split

# Split into features (X) and target (y)
x = df_wine.drop(columns=['quality_label'])  
y = df_wine['quality_label']

# Perform the split (80% training, 20% testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
import pickle

# Save model
with open('wine_quality_model.pkl', 'wb') as f:
    pickle.dump(model, f)

# Load model
with open('wine_quality_model.pkl', 'rb') as f:
    loaded_model = pickle.load(f)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
